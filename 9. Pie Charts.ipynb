### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# Split the column into individual database entries
db_series = df['DatabaseWantToWorkWith'].dropna().str.split(';')

# Flatten the list and count occurrences
all_dbs = db_series.explode()
top_dbs = all_dbs.value_counts().head(5)

# Plot pie chart
plt.figure()
top_dbs.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.ylabel('')  # Hide y-label for cleaner look

plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Split the DevType column into lists
dev_series = df['DevType'].dropna().str.split(';')

# Flatten and count frequencies
all_roles = dev_series.explode()
top_roles = all_roles.value_counts().head(5)

# Plot pie chart
plt.figure()
top_roles.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Developer Roles')
plt.ylabel('')  # cleaner look

plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
# Split the OpSysProfessional use column
os_series = df['OpSysProfessional use'].dropna().str.split(';')

# Flatten and count
all_os = os_series.explode()
top_os = all_os.value_counts().head(5)  # I will view only the top 5 OS

# Plot pie chart
plt.figure()
top_os.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Top Operating Systems Used Professionally')
plt.ylabel('')  # cleaner look

plt.show()
```
