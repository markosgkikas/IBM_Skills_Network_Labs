- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
# KDE plot
sns.kdeplot(
    data=df, 
    x="JobSat", 
    fill=True, 
    clip=(1, 10), 
    bw_adjust=0.8, 
    color="royalblue"
)

# Adding clear ticks for a 1-10 scale
plt.xticks(range(1, 11))
plt.title("Density of Job Satisfaction Scores (1-10)")
plt.xlabel("Satisfaction Level")
plt.grid(axis='y', alpha=0.3)
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# First we need the review the data in these columns
def count_languages(series):
    all_langs = series.dropna().str.split(';')
    counter = Counter()
    for langs in all_langs:
        counter.update(langs)
    return counter

worked = count_languages(df['LanguageHaveWorkedWith'])
wanted = count_languages(df['LanguageWantToWorkWith'])

# Top 10 languages (by worked frequency)
top_langs = [lang for lang, _ in worked.most_common(10)]

worked_counts = [worked[lang] for lang in top_langs]
wanted_counts = [wanted[lang] for lang in top_langs]

# Plot grouped bar chart
x = range(len(top_langs))
width = 0.35

plt.figure()
plt.bar([i - width/2 for i in x], worked_counts, width, label='Worked With')
plt.bar([i + width/2 for i in x], wanted_counts, width, label='Want To Work With')

plt.xticks(x, top_langs, rotation=60)
plt.xlabel('Programming Languages')
plt.ylabel('Frequency')
plt.title('Worked vs Wanted Programming Languages')
plt.legend()
plt.tight_layout()

plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
print('The number of total countries present in the survey is:', df['Country'].unique().shape[0])
```

```python
"""
It is not possible to visualize all 186 countries individually.
We must somehow group these 186 by region, in these case we choose by Continent.
Grouping manually each country to a continent will be time consuming, so we will search online for a py library that does this work.
To complete this task we will use the pycountry library to assign each country in the survey to a continent and then visualize our data.
"""
```

```python
!pip install pycountry-convert
```

```python
import pycountry_convert as pc
```

```python
# Function to map country to continent
def country_to_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_map = {
            'AF': 'Africa',
            'AS': 'Asia',
            'EU': 'Europe',
            'NA': 'North America',
            'SA': 'South America',
            'OC': 'Oceania'
        }
        return continent_map[continent_code]
    except:
        return 'Other'

# Apply mapping
df['Continent'] = df['Country'].apply(country_to_continent)
```

```python
# We replace the 'Hybrid (some remote, some in-person)' with simple 'Hybrid'
df['RemoteWork'] = df['RemoteWork'].replace('Hybrid (some remote, some in-person)', 'Hybrid')
```

```python
heatmap_data = pd.crosstab(df['Continent'], df['RemoteWork'])
```

```python
# Heatmap
plt.figure()
sns.heatmap(heatmap_data, annot=True, fmt='d')

plt.title('Remote Work Distribution by Continent')
plt.xlabel('Remote Work Type')
plt.ylabel('Continent')
plt.tight_layout()

plt.show()
```

```python
# Here we see that Europe dominates in all work types.
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
# We need to replace the strings in the column with values
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0, 
    'More than 50 years': 50})

# Convert to numeric values only
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')

# Drop missing values
df_clean = df[['YearsCodePro', 'JobSat']].dropna()
```

```python
# Pearson Correlation
correlation = df_clean['YearsCodePro'].corr(df_clean['JobSat'])
print("Correlation:", correlation)
```

```python
# As we can see there is weak positive correlation between YearsCodePro and JobSat.
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Create crosstab
df['Employment_clean'] = df['Employment'].replace({
    'Employed, full-time': 'Full-time',
    'Employed, part-time': 'Part-time',
    'Independent contractor, freelancer, or self-employed': 'Self-employed',
    'Student, full-time': 'Student',
    'Student, part-time': 'Student'
})

# We keep only most common Employment categories
top_employment = df['Employment_clean'].value_counts().head(5).index
df_filtered = df[df['Employment_clean'].isin(top_employment)]

# Keep top education levels
top_ed = df_filtered['EdLevel'].value_counts().head(6).index
df_filtered = df_filtered[df_filtered['EdLevel'].isin(top_ed)]

# 2. Crosstab (normalized)
ct = pd.crosstab(
    df_filtered['Employment_clean'],
    df_filtered['EdLevel'],
    normalize='index'
)

# Sort for readability
ct = ct.sort_index()

# 3. Plot stacked bar chart
plt.figure()

ct.plot(
    kind='bar',
    stacked=True
)

plt.title('Employment Status vs Education Level (Distribution)')
plt.xlabel('Employment Status')
plt.ylabel('Proportion')

# Move legend outside
plt.legend(
    title='Education Level',
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.xticks(rotation=30)
plt.tight_layout()

plt.show()
```
