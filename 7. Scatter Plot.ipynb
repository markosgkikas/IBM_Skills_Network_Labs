### Task 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv"

df = pd.read_csv(file_path)
```

### Task 2: Enhancing Scatter Plots


#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
age_code = df[['Age','YearsCodePro']]
age_code = age_code.dropna()

age_code['YearsCodePro'] = pd.to_numeric(age_code['YearsCodePro'], errors='coerce')
age_code = age_code.dropna()
```

```python
plt.scatter(age_code['Age'], age_code['YearsCodePro'])

plt.xlabel('Age')
plt.ylabel('Years of Professional Coding Experience')
plt.title('Age vs Work Experience')

plt.show()
```

```python
# We can also visualize with seaborn for a more clean look while also adding a regression line
```

```python
!pip install seaborn
import seaborn as sns
```

```python
age_code = df[['Age', 'YearsCodePro']].dropna()

# Handle non-numeric entries
age_code['YearsCodePro'] = age_code['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 50
})
age_code['YearsCodePro'] = pd.to_numeric(age_code['YearsCodePro'], errors='coerce')
age_code = age_code.dropna()
```

```python
plt.figure(figsize=(10,6))

sns.set_style("whitegrid")
sns.scatterplot(
    data=age_code,
    x='Age',
    y='YearsCodePro',
    color='blue',
    s=60,
    alpha=0.6
)

plt.title('Age vs Professional Coding Experience', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Years of Professional Coding Experience', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

```python
sns.lmplot(
    data=age_code, x='Age', y='YearsCodePro',
    height=6, aspect=1.5,
    scatter_kws={'s':50, 'alpha':0.6},
    line_kws={'color':'red'}
)
```
