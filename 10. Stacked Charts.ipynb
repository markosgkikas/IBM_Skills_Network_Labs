### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
# Group by Age and calculate median values
grouped = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot stacked bar chart
grouped.plot(kind='bar', stacked=True)

# Labels and title
plt.title('Median Job Satisfaction Points (6 & 7) by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Job Satisfaction Points')
plt.legend(title='Job Satisfaction Metrics')

# Show plot
plt.tight_layout()
plt.show()
```

### Task 2: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
db_col = 'DatabaseWantToWorkWith'

# Split multiple responses (semicolon-separated)
df = df.dropna(subset=[db_col])
df[db_col] = df[db_col].str.split(';')

# Explode into rows
df_exploded = df.explode(db_col)

# Count occurrences per Age Group and Database
grouped = df_exploded.groupby(['Age', db_col]).size().unstack(fill_value=0)

# Keep only top databases overall (to avoid clutter)
top_dbs = grouped.sum().sort_values(ascending=False).head(5).index
grouped = grouped[top_dbs]

# Normalize to proportions
grouped_prop = grouped.div(grouped.sum(axis=1), axis=0)

# Plot stacked chart
grouped_prop.plot(kind='bar', stacked=True)

# Labels and title
plt.title('Preferred Databases to Learn by Age Group (Proportion)')
plt.xlabel('Age Group')
plt.ylabel('Proportion')
plt.legend(title='Database', bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.show()
```

### Task 3: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
# Column with admired languages
lang_col = 'LanguageAdmired'

# Handle multiple responses
df = df.dropna(subset=[lang_col])
df[lang_col] = df[lang_col].str.split(';')

# Explode into rows
df_exploded = df.explode(lang_col)

# Count per Age Group and Language
grouped = df_exploded.groupby(['Age', lang_col]).size().unstack(fill_value=0)

# Keep top languages to avoid clutter
top_langs = grouped.sum().sort_values(ascending=False).head(6).index
grouped = grouped[top_langs]

# Convert to proportions
grouped_prop = grouped.div(grouped.sum(axis=1), axis=0)

# Plot stacked chart
grouped_prop.plot(kind='bar', stacked=True)

# Labels and title
plt.title('Admired Programming Languages by Age Group (Proportion)')
plt.xlabel('Age Group')
plt.ylabel('Proportion')
plt.legend(title='Language', bbox_to_anchor=(1.05, 1))

plt.tight_layout()
plt.show()
```
