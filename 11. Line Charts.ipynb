### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
# Group by Age Group and calculate median compensation
grouped = df.groupby('Age')['ConvertedCompYearly'].median()
grouped = grouped.dropna()

# Ensure proper order of age groups
grouped = grouped.sort_index()

# Plot line chart
plt.figure()
plt.plot(grouped.index.astype(str), grouped.values, marker='o')

# Labels and title
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation (Yearly)')
plt.grid(True)
plt.xticks(rotation=90)

# Improve layout
plt.tight_layout()

# Show plot
plt.show()
```

### Task 2: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
exp_col = 'WorkExp'

# Drop missing values
df = df.dropna(subset=[exp_col, 'ConvertedCompYearly'])

# I will bin experience into ranges for smoother visualization
bins = [0, 2, 5, 10, 15, 20, 30, 50]
labels = ['0-2', '3-5', '6-10', '11-15', '16-20', '21-30', '30+']
df['ExpGroup'] = pd.cut(df[exp_col], bins=bins, labels=labels)

# Group by experience group and compute median compensation
grouped = df.groupby('ExpGroup')['ConvertedCompYearly'].median()

# Ensure correct order
grouped = grouped.sort_index()

# Plot line chart
plt.figure()
plt.plot(grouped.index.astype(str), grouped.values, marker='o')

# Labels and title
plt.title('Median Yearly Compensation by Experience Level')
plt.xlabel('Experience (Years)')
plt.ylabel('Median Converted Compensation (Yearly)')

# Some styling
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
# Use the experience column
exp_col = 'WorkExp'

# Drop missing values
df = df.dropna(subset=[exp_col, 'JobSatPoints_6'])

# Bin experience into groups for smoother trends
bins = [0, 2, 5, 10, 15, 20, 30, 50]
labels = ['0-2', '3-5', '6-10', '11-15', '16-20', '21-30', '30+']
df['ExpGroup'] = pd.cut(df[exp_col], bins=bins, labels=labels)

# Group and compute median job satisfaction
grouped = df.groupby('ExpGroup')['JobSatPoints_6'].median()

# Ensure correct order
grouped = grouped.sort_index()

# Plot line chart
plt.figure()
plt.plot(grouped.index.astype(str), grouped.values, marker='o')

# Labels and title
plt.title('Job Satisfaction (JobSatPoints_6) by Experience Level')
plt.xlabel('Experience (Years)')
plt.ylabel('Median Job Satisfaction')

plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
