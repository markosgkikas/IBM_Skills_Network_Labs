#### Step 2: Load and Preview the Dataset
Load the dataset from the provided URL. Use df.head() to display the first few rows to get an overview of the structure.


```python
# Load the Stack Overflow survey dataset
data_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv"
df = pd.read_csv(data_url)
# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

# Display the first few rows of the dataset
df.head()
```

#### Step 3: Handling Missing Data


Identify and manage missing values in critical columns such as `Employment`, `JobSat`, and `WorkRemote`. Implement a strategy to fill or drop these values, depending on the significance of the missing data.


```python
print('The number of duplicate rows in the dataframe is:',df.duplicated().sum())
```

```python
## Write your code here
print('The number of missing values in the Employment column is:', df['Employment'].isnull().sum())
print('The number of missing values in the JobSat column is:', df['JobSat'].isnull().sum())
print('The number of missing values in the WorkRemote column is:', df['WorkRemote'].isnull().sum())
```

```python
# Since the number of missing rows in Employment column is 0, no further action is needed
# The number of missing rows in the JobSat is 1, since it is categorial it will be filled with the most frequent value
# The number of missing rows in the RemoteWork column is 8, since it is categorial it will be filled with the most frequent value
```

```python
df['JobSat'].value_counts()
```

```python
df['JobSat'] = df['JobSat'].fillna('Slightly satisfied')
```

```python
print('Missing values replaced by: "Slightly satisfied"')
print('The number of missing values in the JobSat column is:', df['JobSat'].isnull().sum())
```

```python
df['JobSat'].value_counts()
```

```python
df['WorkRemote'].value_counts()
```

```python
df['WorkRemote'] = df['WorkRemote'].fillna('Less than once per month / Never')
```

```python
# We check for the change
df['WorkRemote'].value_counts()
```

#### Step 4: Visualize Job Satisfaction


Use a count plot to show the distribution of `JobSat` values. This provides insights into the overall satisfaction levels of respondents.


```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# We use the 'order' parameter to force the sequence defined above
ax = sns.countplot(
    data=df, 
    x='JobSat', 
    palette='viridis')

plt.tight_layout()
```

#### Step 5: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# We use the 'order' parameter to force the sequence defined above
ax = sns.countplot(
    data=df, 
    x='WorkRemote', 
    palette='viridis')

# Rotate the labels 90 degrees
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
df_clean = df.dropna(subset=['WorkRemote', 'Employment']).copy()

# Cross-tabulation: Employment vs RemoteWork
employment_remote_ct = pd.crosstab(df_clean['Employment'], df_clean['WorkRemote'], margins=True)

print("--- Remote Work by Employment Type ---")
print(employment_remote_ct)
```

#### Step 6: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
## Write your code here
# Select relevant columns

df = df[['EdLevel', 'Employment']].dropna()

# Cross-tabulation
crosstab = pd.crosstab(df['EdLevel'], df['Employment'])

print(crosstab)

# Normalize (row-wise %)
crosstab_pct = crosstab.div(crosstab.sum(axis=1), axis=0)

# Stacked Bar Plot
crosstab_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 7)
)
plt.figure(figsize=(12, 8))
plt.title('Employment Type Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion')
plt.legend(title='Employment', bbox_to_anchor=(6, 1))
plt.tight_layout()
plt.show()

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(crosstab_pct, annot=True, fmt='.2f', cmap='YlGnBu')

plt.title('Education vs Employment (Proportion Heatmap)')
plt.xlabel('Employment Type')
plt.ylabel('Education Level')
plt.tight_layout()
plt.show()
```
